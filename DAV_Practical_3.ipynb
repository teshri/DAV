```python
import numpy as np
import pandas as pd
#Creating a dataframe with 3 columns and 50 rows
dat = pd.DataFrame(np.random.rand(50,3), columns=['Col1','Col2','Col3'])
#Replacing 10% values with null values
x = int(0.1*dat.size)
indices_to_replace = np.unravel_index(np.random.choice(dat.size, x, replace=False),[50,3])
dat.iloc[indices_to_replace] = np.nan
#Printing Results
print('Acquired DataFrame:\n')
print(dat)
```

```python
#(a) - Identify and Count missing values
#Identifying missing values
missing_values_identified_df = dat.isnull()
#Counting missing values
total_missing_values = missing_values_identified_df.sum().sum()
#Printing Results
print('DataFrame With Missing Values:\n')
print(missing_values_identified_df)
print('Total Missing Values:', total_missing_values)
```

```python
#(b) - Dropping column with more than 5 null values
#Checking for columns with more than 5 null values
columns_to_drop = dat.columns[dat.isnull().sum()>5]
#Dropping appropriate columns 
columns_dropped_df = dat.drop(columns=columns_to_drop)
#Printing Results 
print('DataFrame With Columns Dropped:\n')
print(columns_dropped_df)
```

```python
#(c) - Identify the row label having maximum of the sum of all values in a row a
#Identifying the row label having max sum
max_sum_row = dat.sum(axis=1).idxmax()
#Dropping appropriate column
max_sum_row_dropped_df = dat.drop(max_sum_row)
#Printing Results
print('Row Index with Max Sum:', max_sum_row, '\n')
print('DataFrame with Row With Max Sum Dropped:\n')
print(max_sum_row_dropped_df)
```

```python
#(d) - Sort the dataframe on the basis of the first column
sorted_df = dat.sort_values(by='Col1')
#Printing Results 
print('DataFrame Sorted by Col1:\n')
print(sorted_df)
```

```python
#(e) - Remove all duplicates from the first column
duplicates_dropped_df = dat.drop_duplicates(subset='Col1')
#Printing Results
print('DataFrame with Duplicates Removed from Col1:\n')
print(duplicates_dropped_df)
```

```python
#(f) - Find the correlation between first and second column and covariance betwe
#Correlation (1st - 2nd col)
corr_1_2 = dat['Col1'].corr(dat['Col2'])
#Covariance (2nd - 3rd col)
cov_2_3 = dat['Col2'].cov(dat['Col3'])
#Printing Results
print('Correlation b/w 1st and 2nd Column:', corr_1_2)
print('Covariance b/w 2nd and 3rd Column:',cov_2_3)
```

```python
#(g) - Detect the outliers and remove the rows having outliers
#Function to Detect and Remove Outliers
def remove_outliers(dat, zscore_thresh=3):
 z_scores = np.abs((dat - dat.mean()) / dat.std())
 outlier_mask = (z_scores < zscore_thresh).all(axis=1)
 new_df = dat[outlier_mask]
 return new_df
#Removing Outliers From Data
outliers_removed_df = remove_outliers(dat)
#Printing Results
print('DataFrame with Outliers Removed:\n')
print(outliers_removed_df)
```

```python
#(h) - Discretize second column and create 5 bins
#Filling Missing (NaN) Values in Col2 because pandas.cut cannot appropriately ha
dat['Col2'] = dat['Col2'].fillna(dat['Col2'].mean())
#Envoking pandas.cut funtion 
dat['Col2_Discretized'] = pd.cut(dat.iloc[:,1], bins=5, labels=False)
#Printing Results
print('DataFrame with Col2 Discretized into 5 bins:\n')
print(dat)
```
